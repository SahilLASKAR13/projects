```python
path=r"G:\corizo data science internship\project minor\spotify dataset.csv"
```

```python
df_all=pd.read_csv(path,usecols=['track_name','track_artist','track_popularity','track_album_name','playlist_name','playlist_genre'])
df_all
```

```python
x=df_all.iloc[:,0:5]
x
```

```python
print(df_all["track_name"].unique())
print(df_all["track_artist"].unique())
print(df_all["track_album_name"].unique())
print(df_all["playlist_name"].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
l_tn=LabelEncoder()
l_ta=LabelEncoder()
l_tan=LabelEncoder()
l_pop=LabelEncoder()
l_pl=LabelEncoder()
```

```python
x['new_tn']=l_tn.fit_transform(x["track_name"])
x['new_ta']=l_ta.fit_transform(x["track_artist"])
x['new_tan']=l_tan.fit_transform(x["track_album_name"])
x['new_pop']=l_pop.fit_transform(x["track_popularity"])
x['new_pl']=l_pop.fit_transform(x["playlist_name"])
```

```python
x
```

```python
x2=x.iloc[:,5:10]
x2
```

```python
y=df_all.iloc[:,5:]
y
```

```python
l_y=LabelEncoder()
```

```python
y['new_y']=l_y.fit_transform(y["playlist_genre"])
```

```python
y
```

```python
y2=y.iloc[:,1:]
y2
```

```python
y2["new_y"].unique()
```

```python
sns.pairplot(df_all,hue="playlist_genre")
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x2,y2,train_size=0.8)
```

```python
print(len(xtrain))
print(len(ytrain))
print(len(xtest))
print(len(ytest))
```

Support Vector Machine (Classifier)

```python
from sklearn.svm import SVC
svc_model=SVC()
svc_model.fit(xtrain,ytrain)
```

```python
ans=svc_model.predict(xtest)
ans
```

```python
ytest
```

```python
svc_model.score(xtest,ytest)
```

DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

```python
ans2=dtc.predict(xtest)
ans2
```

```python
dtc.score(xtest,ytest)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(xtrain,ytrain)
```

```python
ans3=rfc.predict(xtest)
ans3
```

```python
print(rfc.score(xtest,ytest))
```

Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
m=GaussianNB()
m.fit(xtrain,ytrain)
```

```python
ans4=m.predict(xtest)
ans4
```

```python
print(m.score(xtest,ytest))
```
